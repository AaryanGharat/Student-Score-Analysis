```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pwd = os.getcwd()
filepath=os.path.join(pwd,"dataset/Expanded_data_with_more_features.csv").replace("\\", "/")
```

```python
df=pd.read_csv(filepath)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Drop Unnamed column

```python
df=df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data = df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

**FROM ABOVE CHART WE ANALYSE THAT NO. OF FEMALES ARE MORE THAN MALES**

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between parents Education and Student Score")
plt.show()
```

**From the above chart we have concluded that the parents education have good impact on marks**

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parents Marital Status and Student Score")
plt.show()
```

**From the above chart we have concluded that the parents marital status have no or negligible impact on score**

**Distribution of ethinic groups**

```python
df["EthnicGroup"].unique()
```

```python
groupA= df.loc[(df["EthnicGroup"]=="group A")].count()
groupB= df.loc[(df["EthnicGroup"]=="group B")].count()
groupC= df.loc[(df["EthnicGroup"]=="group C")].count()
groupD= df.loc[(df["EthnicGroup"]=="group D")].count()
groupE= df.loc[(df["EthnicGroup"]=="group E")].count()
l=['group A' ,'group B','group C' , 'group D', 'group E']
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels=l,autopct="%1.2f%%")
plt.title("Distribution of ethinic groups")
plt.show()
```
